```python
import sqlite3
import pandas as pd

# Create a new SQL database in memory
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
# Create table
cursor.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    marks INTEGER,
    gender TEXT,
    city TEXT
)
""")
cursor.executemany("""
INSERT INTO students (name, marks, gender, city)
VALUES (?, ?, ?, ?)
""", [
    ("Amit", 85, "Male", "Delhi"),
    ("Riya", 92, "Female", "Mumbai"),
    ("Sourav", 76, "Male", "Kolkata"),
    ("Anita", 89, "Female", "Chennai"),
    ("Rahul", 60, "Male", "Pune")
])
conn.commit()
df = pd.read_sql_query("SELECT * FROM students", conn)
df.head()
```

```python
!pip install requests beautifulsoup4
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send request to the website
url = "https://quotes.toscrape.com/"
response = requests.get(url)

# Step 2: Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Step 3: Extract data (quotes + authors + tags)
quotes_data = []

for quote in soup.find_all("div", class_="quote"):
    text = quote.find("span", class_="text").get_text()
    author = quote.find("small", class_="author").get_text()
    tags = [tag.get_text() for tag in quote.find_all("a", class_="tag")]

    quotes_data.append({
        "Quote": text,
        "Author": author,
        "Tags": ", ".join(tags)
    })

# Step 4: Convert to DataFrame
df = pd.DataFrame(quotes_data)

# Show scraped data
df.head()

```

```python
!pip install selenium webdriver-manager

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time
import matplotlib.pyplot as plt

# Step 1: Start Browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Step 2: Open Website
query = "laptops"
url = f"https://www.flipkart.com/search?q={query}"
driver.get(url)
time.sleep(3)

# Step 3: Scrape Data
titles = driver.find_elements(By.CSS_SELECTOR, "div._4rR01T")
prices = driver.find_elements(By.CSS_SELECTOR, "div._30jeq3")

data = []
for t, p in zip(titles, prices):
    data.append({
        "Product Title": t.text,
        "Price": p.text
    })

df = pd.DataFrame(data)

# Step 4: Close browser
driver.quit()

# Step 5: Show Data
df.head()

# Step 6: Visualization
plt.figure(figsize=(10,6))
plt.barh(df['Product Title'][:10], 
         df['Price'].str.replace("₹","").str.replace(",","").astype(int)[:10])
plt.title("Top 10 Laptop Prices from Flipkart")
plt.xlabel("Price (₹)")
plt.show()
```

```python
# -----------------------------------------------------
# LOAD AUDIO FILE (MP3/WAV) IN JUPYTER NOTEBOOK
# -----------------------------------------------------

# Install required libraries (run only once)
!pip install pydub librosa soundfile

from tkinter import Tk
from tkinter.filedialog import askopenfilename
from pydub import AudioSegment
from IPython.display import Audio
import librosa
import numpy as np

# -------------------------------------
# Step 1: Open file selection dialog
# -------------------------------------
Tk().withdraw()
file_path = askopenfilename(title="Select an audio file (MP3/WAV)")

if not file_path:
    print("No file selected.")
else:
    print("Selected:", file_path)

    # -------------------------------------
    # Step 2: Load audio using pydub
    # -------------------------------------
    if file_path.endswith(".mp3"):
        audio = AudioSegment.from_mp3(file_path)
    elif file_path.endswith(".wav"):
        audio = AudioSegment.from_wav(file_path)
    else:
        print("Unsupported format")
        audio = None

    # -------------------------------------
    # Step 3: Play audio in Jupyter
    # -------------------------------------
    if audio:
        samples = np.array(audio.get_array_of_samples())
        samplerate = audio.frame_rate

        print("Audio Duration (seconds):", len(audio) / 1000)
        print("Sample Rate:", samplerate)

        # Jupyter inline audio player
        display(Audio(samples, rate=samplerate))

        # -------------------------------------
        # Step 4: Load using librosa (for ML work)
        # -------------------------------------
        y, sr = librosa.load(file_path, sr=None)
        print("Librosa Loaded Shape:", y.shape)
        print("Librosa Sample Rate:", sr)
```
