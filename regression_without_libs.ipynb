```python
# linear regression
x=[1,2,3,4,5]
y=[2,4,5,4,5]
def linear_regression(x,y):
    n=len(x)
    total_x=sum(x)
    total_y=sum(y)
    total_xy=sum([x[i]*y[i] for i in range(n)])
    total_x2=sum([x[i]**2 for i in range(n)])
    m=(n*total_xy-total_x*total_y)/(n*total_x2-total_x**2)
    c=(total_y-m*total_x)/n
    return m,c
def predict(x_value,m,c):
    return m*x_value+c
m,c=linear_regression(x,y)
print("Slope (m):",m)
print("Intercept (c):",c)
x_new=6
predicted_y=predict(x_new,m,c)
print(f"When x = {x_new}, predicted y = {predicted_y}")
```
